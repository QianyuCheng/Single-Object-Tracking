```python
models = ['pf', 'ssvm', 'kcf']

for video_idx in range(50):
    plt.figure(figsize=[6, 4])
    for model_name in models:
        df = pd.read_csv('{}.log'.format(model_name), sep='\t', header=None)
        plt.plot(df.iloc[video_idx, 2:], label=model_name.upper())
    plt.title(df.iloc[video_idx, 0])
    plt.legend(loc='best')
    plt.xlabel('Frames')
    plt.ylabel('RMSE')
    plt.savefig('results/{}.jpg'.format(df.iloc[video_idx, 0]))
    plt.show()
```
